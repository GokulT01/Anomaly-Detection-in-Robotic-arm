```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, GRU, RepeatVector, TimeDistributed, Dense
from tensorflow.keras.optimizers import Adam
import time
import warnings
warnings.filterwarnings("ignore")

# ---------- 1. Load and preprocess training & test data ----------
try:
    df_train = pd.read_csv('robot_data_cleaned_properly.csv')
    df_test = pd.read_csv('test_data.csv')
except Exception as e:
    print(f"Error loading data: {e}")
    raise

# Compute joint errors
for df in [df_train, df_test]:
    for j in range(1, 7):
        df[f'ROBOT_TARGET_JOINT_POSITIONS (J{j})'] = pd.to_numeric(df[f'ROBOT_TARGET_JOINT_POSITIONS (J{j})'], errors='coerce')
        df[f'ROBOT_ACTUAL_JOINT_POSITIONS (J{j})'] = pd.to_numeric(df[f'ROBOT_ACTUAL_JOINT_POSITIONS (J{j})'], errors='coerce')
        df[f'JOINT_ERROR_J{j}'] = abs(df[f'ROBOT_TARGET_JOINT_POSITIONS (J{j})'] - df[f'ROBOT_ACTUAL_JOINT_POSITIONS (J{j})'])

# ---------- 2. Preprocessing function ----------
def preprocess_features(df, joint_num):
    features = [
        f'JOINT_ERROR_J{joint_num}',
        f'ROBOT_TARGET_JOINT_VELOCITIES (J{joint_num})',
        f'ROBOT_ACTUAL_JOINT_VELOCITIES (J{joint_num})',
        f'ROBOT_TARGET_JOITN_CURRENT (J{joint_num})',
        f'ROBOT_ACTUAL_JOINT_CURRENT (J{joint_num})'
    ]
    data = df[features].copy()
    for col in features[1:]:
        data[col] = data[col].astype(str).str.replace('[', '', regex=False).str.replace(']', '', regex=False)
        data[col] = pd.to_numeric(data[col].str.strip(), errors='coerce')
    data = data.fillna(method='ffill')
    scaler = MinMaxScaler()
    scaled = scaler.fit_transform(data)
    return scaled, scaler

# ---------- 3. Autoencoder Trainers ----------
def train_autoencoder(model_type, X_train, seq_len):
    X_seq = np.array([X_train[i:i+seq_len] for i in range(len(X_train)-seq_len)])
    if model_type == 'LSTM':
        model = Sequential([
            LSTM(64, activation='relu', input_shape=(seq_len, X_train.shape[1]), return_sequences=False),
            RepeatVector(seq_len),
            LSTM(64, activation='relu', return_sequences=True),
            TimeDistributed(Dense(X_train.shape[1]))
        ])
    elif model_type == 'GRU':
        model = Sequential([
            GRU(64, activation='relu', input_shape=(seq_len, X_train.shape[1]), return_sequences=False),
            RepeatVector(seq_len),
            GRU(64, activation='relu', return_sequences=True),
            TimeDistributed(Dense(X_train.shape[1]))
        ])
    model.compile(optimizer=Adam(0.001), loss='mse')
    start_time = time.time()
    model.fit(X_seq, X_seq, epochs=10, batch_size=32, verbose=0)
    elapsed_time = time.time() - start_time
    return model, elapsed_time

# ---------- 4. Anomaly Detection Function ----------
def detect_anomalies(model, X_test, seq_len, df_test, joint_num, label='Original'):
    X_seq = np.array([X_test[i:i+seq_len] for i in range(len(X_test)-seq_len)])
    preds = model.predict(X_seq)
    err = np.mean(np.square(X_seq - preds), axis=(1,2))
    threshold = np.percentile(err, 99.9)
    anomalies = np.where(err > threshold)[0] + seq_len

    def group_indices(indices):
        indices = sorted(indices)
        groups = []
        if not indices:
            return groups
        start = end = indices[0]
        for idx in indices[1:]:
            if idx == end + 1:
                end = idx
            else:
                groups.append((start, end))
                start = end = idx
        groups.append((start, end))
        return groups

    grouped = group_indices(anomalies.tolist())
    print(f"\n{label} Detection ‚Äì Joint {joint_num} Anomaly Intervals:")
    for start, end in grouped:
        values = df_test.iloc[start:end+1][f'JOINT_ERROR_J{joint_num}']
        print(f"  Interval {start}-{end}: Max error = {values.max():.5f}, Avg error = {values.mean():.5f}")

    plt.figure(figsize=(10,4))
    plt.hist(err, bins=100, alpha=0.7)
    plt.axvline(threshold, color='red', linestyle='--', label='Threshold')
    plt.title(f'Reconstruction Error Distribution - Joint {joint_num} ({label})')
    plt.xlabel("Reconstruction Error")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid(True)
    plt.show()

    plt.figure(figsize=(14,5))
    plt.plot(df_test[f'JOINT_ERROR_J{joint_num}'], label=f'Joint {joint_num} Error')
    for start, end in grouped:
        plt.axvspan(start, end, color='red', alpha=0.3)
    plt.title(f'Anomaly Detection on Joint {joint_num} ({label})')
    plt.xlabel('Index')
    plt.ylabel('Error')
    plt.legend()
    plt.grid(True)
    plt.show()

    return anomalies.tolist(), err, threshold

# ---------- 5. Summarize Anomalies ----------
def summarize_anomalies(model_name, anomaly_indices, reconstruction_errors, threshold, joint_num, elapsed_time):
    def group_indices(indices):
        indices = sorted(indices)
        groups = []
        if not indices:
            return groups
        start = end = indices[0]
        for idx in indices[1:]:
            if idx == end + 1:
                end = idx
            else:
                groups.append((start, end))
                start = end = idx
        groups.append((start, end))
        return groups

    groups = group_indices(anomaly_indices)
    num_intervals = len(groups)
    interval_lengths = [end - start + 1 for start, end in groups]

    if len(anomaly_indices) > 0:
        max_error = max([reconstruction_errors[i] for i in anomaly_indices if i < len(reconstruction_errors)])
        mean_error = np.mean([reconstruction_errors[i] for i in anomaly_indices if i < len(reconstruction_errors)])
    else:
        max_error = 0
        mean_error = 0

    return {
        "Model": model_name,
        "Joint": joint_num,
        "Anomaly Intervals": num_intervals,
        "Avg Interval Length": np.mean(interval_lengths) if interval_lengths else 0,
        "Max Interval Length": max(interval_lengths) if interval_lengths else 0,
        "Max Error in Anomalies": max_error,
        "Mean Error in Anomalies": mean_error,
        "Threshold Used": threshold,
        "Train Time (s)": elapsed_time
    }

# ---------- 6. Run model analysis on all joints for both LSTM and GRU ----------
seq_len = 50
all_summaries = []

for model_type in ['LSTM', 'GRU']:
    print(f"\n================= Running {model_type} Autoencoder =================")
    for joint in range(1, 7):
        print(f"\nüöÄ Running {model_type} model on Joint {joint} (Before Injection)...")
        X_train, scaler = preprocess_features(df_train, joint)
        X_test, _ = preprocess_features(df_test, joint)
        model, elapsed = train_autoencoder(model_type, X_train, seq_len)
        anomalies, err, threshold = detect_anomalies(model, X_test, seq_len, df_test, joint, label='Original')
        summary = summarize_anomalies(model_type, anomalies, err, threshold, joint, elapsed)
        all_summaries.append(summary)

    if model_type == 'LSTM':
        print("\nüîß Injecting synthetic anomaly into Joint 3 from index 10000 to 10010")
        df_test.loc[10000:10010, 'JOINT_ERROR_J3'] += 0.05

    print(f"\nüîÅ Re-running {model_type} model on Joint 3 (After Injection)...")
    X_train, scaler = preprocess_features(df_train, 3)
    X_test, _ = preprocess_features(df_test, 3)
    model, elapsed = train_autoencoder(model_type, X_train, seq_len)
    anomalies, err, threshold = detect_anomalies(model, X_test, seq_len, df_test, 3, label='With Injection')
    summary = summarize_anomalies(model_type + " (Injected)", anomalies, err, threshold, 3, elapsed)
    all_summaries.append(summary)

# ---------- 7. Display Overall Comparison ----------
comparison_df = pd.DataFrame(all_summaries)
print("\n================= Overall Model Comparison Summary =================")
print(comparison_df)

# Aggregate comparison per model type (excluding Injected runs)
overall_stats = comparison_df[~comparison_df['Model'].str.contains("Injected")].groupby('Model').agg({
    'Anomaly Intervals': 'sum',
    'Avg Interval Length': 'mean',
    'Max Interval Length': 'max',
    'Max Error in Anomalies': 'mean',
    'Mean Error in Anomalies': 'mean',
    'Threshold Used': 'mean',
    'Train Time (s)': 'mean'
}).reset_index()

print("\n================= Aggregated Model Performance =================")
print(overall_stats)

# Visual comparison
plt.figure(figsize=(12, 6))
bar_width = 0.25
x = np.arange(len(overall_stats['Model']))

plt.bar(x - bar_width, overall_stats['Max Error in Anomalies'], width=bar_width, label='Max Error')
plt.bar(x, overall_stats['Mean Error in Anomalies'], width=bar_width, label='Mean Error')
plt.bar(x + bar_width, overall_stats['Train Time (s)'], width=bar_width, label='Avg Train Time (s)')

plt.xticks(x, overall_stats['Model'])
plt.ylabel('Metric Value')
plt.title('LSTM vs GRU - Error and Time Comparison (All Joints)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
def evaluate_joint_health(df, joint_num):
    error_col = f'JOINT_ERROR_J{joint_num}'
    errors = df[error_col].dropna()

    max_err = errors.max()
    avg_err = errors.mean()

    if max_err >= 0.45:
        return f"Joint {joint_num}: CRITICAL ‚Äì max error {max_err:.5f} exceeds 0.45 threshold"
    elif max_err >= 0.1:
        return f"Joint {joint_num}: WARNING ‚Äì max error {max_err:.5f} exceeds 0.1 threshold"
    elif avg_err >= 0.01:
        return f"Joint {joint_num}: MONITOR ‚Äì avg error {avg_err:.5f} suggests potential early warning"
    else:
        return f"Joint {joint_num}: GOOD ‚Äì operating within safe error bounds"

print("\n================= FINAL JOINT HEALTH STATUS =================")
for j in range(1, 7):
    print(evaluate_joint_health(df_test, j))
```
